In Python, class methods and static methods are two types of methods that are associated with the class itself rather than with an instance of the class.

**Class Method**

A class method is a method that is bound to the class and not the instance of the class. It takes the class itself as its first argument (usually named cls), rather than the instance (self). This allows the method to access or modify class-level attributes, rather than instance-level attributes.

Class methods are defined using the @classmethod decorator.

**Static Method**

A static method is a method that does not take the instance (self) or the class (cls) as its first argument. It behaves like a regular function that belongs to the class's namespace. Static methods cannot access or modify class or instance attributes. They are defined using the @staticmethod decorator.

```python
#answer 2

from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Derived class for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Derived class for Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage
if __name__ == "__main__":
    circle = Circle(5)  # Circle with radius 5
    rectangle = Rectangle(4, 6)  # Rectangle with width 4 and height 6

    print(f"Area of the circle: {circle.area():.2f}")
    print(f"Area of the rectangle: {rectangle.area()}")
```

```python
#answer 3

# Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def display_type(self):
        print(f"Vehicle Type: {self.vehicle_type}")

# Derived class
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def display_info(self):
        print(f"Car Brand: {self.brand}")

# Further derived class
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery_capacity = battery_capacity

    def display_battery_info(self):
        print(f"Battery Capacity: {self.battery_capacity} kWh")

# Example usage
if __name__ == "__main__":
    # Create an ElectricCar object
    tesla = ElectricCar("Electric Vehicle", "Tesla", 100)

    # Display information
    tesla.display_type()            # From Vehicle
    tesla.display_info()            # From Car
    tesla.display_battery_info()    # From ElectricCar
```

```python
#answer 5

class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.__balance = initial_balance  # Private attribute
        self.account_holder = account_holder

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        elif amount > self.__balance:
            print("Insufficient balance.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to check balance
    def check_balance(self):
        print(f"Current Balance: ${self.__balance:.2f}")

# Example usage
if __name__ == "__main__":
    # Create a bank account
    account = BankAccount("John Doe", 1000)

    # Perform transactions
    account.check_balance()  # Check initial balance
    account.deposit(500)     # Deposit money
    account.withdraw(300)    # Withdraw money
    account.withdraw(1500)   # Attempt to withdraw more than balance
    account.check_balance()  # Check final balance
```

```python
#answer 6

# Base class
class Instrument:
    def play(self):
        raise NotImplementedError("Subclass must implement abstract method")

# Derived class for Guitar
class Guitar(Instrument):
    def play(self):
        print("Playing the guitar: Strumming the strings!")

# Derived class for Piano
class Piano(Instrument):
    def play(self):
        print("Playing the piano: Pressing the keys!")

# Function to demonstrate polymorphism
def perform(instrument):
    instrument.play()

# Example usage
if __name__ == "__main__":
    # Create instances of Guitar and Piano
    guitar = Guitar()
    piano = Piano()

    # Demonstrate runtime polymorphism
    print("Performance with Guitar:")
    perform(guitar)  # Calls Guitar's play method

    print("\nPerformance with Piano:")
    perform(piano)  # Calls Piano's play method
```

```python
#answer 7

class MathOperations:
    @classmethod
    def add_numbers(cls, num1, num2):
        """Class method to add two numbers."""
        return num1 + num2

    @staticmethod
    def subtract_numbers(num1, num2):
        """Static method to subtract two numbers."""
        return num1 - num2

# Example usage
if __name__ == "__main__":
    # Using the class method to add numbers
    result_add = MathOperations.add_numbers(10, 5)
    print(f"Addition Result: {result_add}")

    # Using the static method to subtract numbers
    result_subtract = MathOperations.subtract_numbers(10, 5)
    print(f"Subtraction Result: {result_subtract}")
```

```python
#answer 9

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        """Override the str method to display the fraction."""
        return f"{self.numerator}/{self.denominator}"

# Example usage
if __name__ == "__main__":
    # Create a Fraction object
    fraction1 = Fraction(3, 4)
    fraction2 = Fraction(5, 8)

    # Print the fractions
    print(f"Fraction 1: {fraction1}")
    print(f"Fraction 2: {fraction2}")
```

```python
#answer 10

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        """Override the + operator to add two vectors."""
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        raise TypeError("Operand must be of type Vector")

    def __str__(self):
        """String representation of the vector."""
        return f"Vector({self.x}, {self.y})"

# Example usage
if __name__ == "__main__":
    # Create two Vector objects
    vector1 = Vector(2, 3)
    vector2 = Vector(4, 5)

    # Add the two vectors using the + operator
    result = vector1 + vector2

    # Display the result
    print(f"Vector 1: {vector1}")
    print(f"Vector 2: {vector2}")
    print(f"Result of addition: {result}")
```

```python
#answer 12

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        """Compute and return the average of the grades."""
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

# Example usage
if __name__ == "__main__":
    # Create a Student object
    student = Student("John Doe", [85, 90, 78, 92, 88])

    # Calculate and print the average grade
    print(f"Average grade for {student.name}: {student.average_grade():.2f}")
```

```python
#answer 14

# Base class
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        """Calculate the salary based on hours worked and hourly rate."""
        return self.hours_worked * self.hourly_rate

# Derived class
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        """Calculate the salary with a bonus for the manager."""
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example usage
if __name__ == "__main__":
    # Create an Employee object
    employee = Employee("John Doe", 160, 20)

    # Create a Manager object
    manager = Manager("Alice Smith", 160, 30, 500)

    # Calculate and print the salary for the employee and manager
    print(f"{employee.name}'s Salary: ${employee.calculate_salary()}")
    print(f"{manager.name}'s Salary (with bonus): ${manager.calculate_salary()}")
```

```python
#answer 16

from abc import ABC, abstractmethod

# Base class with abstract method
class Animal(ABC):
    @abstractmethod
    def sound(self):
        """Abstract method to be implemented by derived classes."""
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        """Implement the sound for Cow."""
        return "Moo"

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        """Implement the sound for Sheep."""
        return "Baa"

# Example usage
if __name__ == "__main__":
    # Create instances of Cow and Sheep
    cow = Cow()
    sheep = Sheep()

    # Call the sound method for both
    print(f"Cow sound: {cow.sound()}")
    print(f"Sheep sound: {sheep.sound()}")
```

```python
#answer 18

# Base class House
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def get_details(self):
        """Return the details of the house."""
        return f"Address: {self.address}\nPrice: ${self.price}"

# Derived class Mansion
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def get_details(self):
        """Return the details of the mansion, including the number of rooms."""
        base_details = super().get_details()
        return f"{base_details}\nNumber of Rooms: {self.number_of_rooms}"

# Example usage
if __name__ == "__main__":
    # Create a House object
    house = House("123 Elm Street", 250000)

    # Create a Mansion object
    mansion = Mansion("456 Oak Avenue", 5000000, 12)

    # Get and print the details of the house and mansion
    print("House Details:")
    print(house.get_details())

    print("\nMansion Details:")
    print(mansion.get_details())
```
